```python
import csv
import matplotlib.pyplot as plt
```

세종시 최고 강수량

```python
file = open('/content/rain.csv', 'r', encoding = 'cp949')
saejong_data = csv.reader(file)
header = next(saejong_data)

max_rain = -999
max_date = ''

for row in saejong_data:
  #최고 강수량 데이터가 ''(결측치) 인지 확인
  if row[-1] == '':
    row[-1] = -999 # -999 넣어서 결측치가 있떤 자리라고 표시
  if row[-1] == str(0):
    row[-1] = -999
    
  row[-1] = float(row[-1])
  # print(row[0], row[-1])    #날짜 데이터와 같이 출력

  if max_rain < row[-1]:
    max_date = row[0]
    max_rain = row[-1]

file.close()

print('세종시 강수량 관측 이래 최대 강수량은',max_rain, 'mm이며, 해당 날짜는', max_date,'입니다.')
```

세종시 최저 강수량

```python
file = open('/content/rain.csv', 'r', encoding = 'cp949')
saejong_data = csv.reader(file)
header = next(saejong_data)

min_rain = 999
min_date = ''

for row in saejong_data:
  #최저 강수량 데이터가 ''(결측치) 인지 확인
  if row[-1] == '':
    row[-1] = 999 # 999 넣어서 결측치가 있떤 자리라고 표시
  if row[-1] == str(0):
    row[-1] = 999
    
  row[-1] = float(row[-1])
  # print(row[0], row[-1])    #날짜 데이터와 같이 출력

  if min_rain > row[-1]:
    min_date = row[0]
    min_rain = row[-1]

file.close()

print('세종시 강수량 관측 이래 최소 강수량은',min_rain, 'mm이며, 해당 날짜는', min_date,'입니다.')
```

```python
file = open('/content/rain.csv', 'r', encoding = 'cp949')
saejong_data = csv.reader(file)
header = next(saejong_data)

result = []   # 결측치를 제거한 강수량 리스트

for row in saejong_data:
  if row[-1] != '':   # 강수량이 ''(결측치)가 아니면, result에 float(row[-1]) 담아라
    result.append(float(row[-1]))

plt.title('Precipitation')
plt.plot(result, 'b')
plt.show()
```

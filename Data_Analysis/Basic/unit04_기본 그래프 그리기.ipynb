plt.plot(데이터셋) 그래프 그리기

```python
import matplotlib.pyplot as plt   #맷플로립 라이브러리 임포트
```

```python
plt.plot([10, 20, 30, 40, 10, 20, 20, 50, 80, 40, 10])  # 주어진 값은 y값이다. x값은 자동으로 0부터 1씩 증가함
plt.show()
```

plt.plot(x, y) 그래프 그리기

```python
plt.plot(['start', 2, 3, 4, 5, 6, 7, 8, 9, 10, 'end'], [10, 20, 30, 40, 10, 20, 20, 50, 80, 40, 10])  # ([x], [y])
plt.show()
```

```python
plt.title('plotting') # 타이틀
plt.plot(['start', 2, 3, 4, 5, 6, 7, 8, 9, 10, 'end'], [10, 20, 30, 40, 10, 20, 20, 50, 80, 40, 10])  # ([x], [y])
plt.show()
```

```python
plt.title('plotting') # 타이틀
plt.plot([10, 20, 30, 40, 10, 20, 20, 50, 80, 40, 10], label = 'asc') # 증가를 의미하는 asc 범례 추가
plt.plot([40, 30, 20, 10, 70, 80, 20, 40 ,30, 40, 50], label = 'desc')  # 감소를 의미하는 desc 범례 추가

plt.legend(loc = 1)  #그래프에 범례 속성 추가 (label = 'asc' or 'desc') ==== loc 옵션(범례 위치) 디폴트는 5
plt.show()
```

```python
plt.title('plotting') # 타이틀
plt.plot([10, 20, 30, 40, 10, 20, 20, 50, 80, 40, 10], color = 'skyblue', label = 'skyblue') # 증가를 의미하는 asc 범례 추가
plt.plot([40, 30, 20, 10, 70, 80, 20, 40 ,30, 40, 50], 'pink', label = 'pink')  # 감소를 의미하는 desc 범례 추가
                                                                                    #color = 선 색 ...컬러 = 'ㅁㄴㅇㄹ' or 'ㅁㄴㅇㄹ'
plt.legend(loc = 1)  #그래프에 범례 속성 추가 (label = 'asc' or 'desc') .... loc 옵션(범례 위치) 디폴트는 5
plt.show()
```

```python
plt.title('plotting') # 타이틀
plt.plot([10, 20, 30, 40, 10, 20, 20, 50, 80, 40, 10], color = 'r', label = 'red') # 기존 색상은 약자로 가능
plt.plot([40, 30, 20, 10, 70, 80, 20, 40 ,30, 40, 50], 'k', label = 'black')  #검정은 k
plt.legend(loc = 1)
plt.show()
```

```python
plt.title('linestyle')

plt.plot([10, 20, 30, 40, 10, 20, 20, 50, 80, 40, 10], color = 'r', linestyle = '--', label = 'red') # 빨간 dashed 그래프
plt.plot([40, 30, 20, 10, 70, 80, 20, 40 ,30, 40, 50], 'k', ls = ':', label = 'black')  # 검정 dotted 그래프 ...ls로도 가능
plt.legend(loc = 1)
plt.show()
```

```python
plt.title('linestyle')

plt.plot([10, 20, 30, 40, 10, 20, 20, 50, 80, 40, 10], 'r.', label = 'red') # 빨간 dashed 그래프...color 생략 가능
plt.plot([40, 30, 20, 10, 70, 80, 20, 40 ,30, 40, 50], 'k^', label = 'black')  # 검정 dotted 그래프 ...ls로도 가능
plt.legend(loc = 1)
plt.show()
```

```python
plt.title('linestyle')

plt.plot([10, 20, 30, 40, 10, 20, 20, 50, 80, 40, 10], 'r.', label = 'red') # 빨간 dashed 그래프
plt.plot([10, 20, 30, 40, 10, 20, 20, 50, 80, 40, 10], color = 'g') 
plt.plot([40, 30, 20, 10, 70, 80, 20, 40 ,30, 40, 50], 'k^--', label = 'black')  # 검정 dotted 그래프 ...ls로도 가능
                                                        # 선까지...하지만 색을 다르게는 못함
plt.legend(loc = 1)
plt.show()
```

```python
plt.title('marker')

plt.plot([10, 20, 30, 40, 10, 20, 20, 50, 80, 40, 10], 'r.-', label = 'circle') # 빨간 dashed 그래프
plt.plot([40, 30, 20, 10, 70, 80, 20, 40 ,30, 40, 50], 'k^--', label = 'black')  # 검정 dotted 그래프 ...ls로도 가능
                                                        # 선까지...하지만 색을 다르게는 못함
plt.legend(loc = 1)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

file = open('/content/seoul.csv', 'r', encoding = 'cp949')
data = csv.reader(file)
header = next(data)

result = []   # 결측치를 제거한 최고 기온 리스트

for row in data:
  if row[-1] != '':   # 최고 기온이 ''(결측치)가 아니면, result에 float(row[-1]) 담아라
    result.append(float(row[-1]))

plt.figure(figsize = (10, 2), dpi = 200)   #해상도 등 수정
plt.plot(result, 'r')
plt.show()
```

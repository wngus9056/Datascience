1. 이진 분류 데이터셋(forge) 확인하기

데이터셋 다운로드

```python
import mglearn
import matplotlib.pyplot as plt

# 데이터셋 만들기
X, y = mglearn.datasets.make_forge()
```

데이터 확인하기

```python
print('X.shape :', X.shape)
print('y.shape :', y.shape)
```

```python
X
```

```python
plt.figure(dpi = 120)
plt.rc('font', family = 'NanumBarunGothic')

# 산점도 그리기
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)

plt.legend(['클래스 0', '클래스 1'], loc = 4)
plt.xlabel('첫 번째 특성')
plt.ylabel('두 번째 특성')
plt.show()
```

분류 문제정의 : 위스콘신 유방암 데이터 셋을 사용한 악성 종양(Label, 1) 예측하기

데이터 준비하기 (https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic))

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
print(cancer.keys())  # 데이터셋 키 확인
```

데이터 확인하기

```python
cancer.data.shape # 유방암 데이터의 형태 -> 569건, 특성 30개
```

```python
# 클래스별 샘플 갯수
import numpy

# 양성, 악성 종양의 비율 37% : 63%
{n : v for n, v in zip(cancer.target_names, numpy.bincount(cancer.target))}
```

```python
print(cancer.keys())
```

```python
# 데이터의 특성 (feature) 확인
cancer.feature_names
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    cancer['data'], cancer['target'],  # 문제지와 답을 주고
    test_size = 0.25, random_state = 777) # 랜덤을 안주면 매번 할 때마다 새로 뽑아서 결과가 매번 바뀐다.

import matplotlib.pyplot as plt
import pandas

cancer_df = pandas.DataFrame(X_train, columns = cancer.feature_names)

cancer_df.head()
```

```python
# 데이터프레임을 사용해 y_train에 따라 색으로 구분된 산점도 행렬을 만든다.

pandas.plotting.scatter_matrix(cancer_df, c = y_train, figsize = (15, 15),
                               marker = 'o', hist_kwds = {'bins' : 20}, s = 60, alpha =.8)

plt.show()

plt.imshow([numpy.unique(y_train)])  # 밑에 범례같은거 추가
_ = plt.xticks(ticks=numpy.unique(y_train), labels = cancer['feature_names']) # _는 없어도 상관은 없지만 return값을 왜 안받냐고 화낼 때가 있다.
```

데이터 준비하기

```python
import csv
import matplotlib.pyplot as plt
```

```python
# 최고 기온 데이터만 리스트에 담아서 출력하기

f = open('/content/seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

result = [] # 최고 기온 데이터를 저장할 리스트 생성

for row in data:
  if row[-1] != '':   # 최고 기온 데이터 값이 존재하면 (결측치 제외)
    result.append(float(row[-1]))

print(result) # 최고 기온 리스트
```

데이터 개수 확인

```python
print(len(result))    # 결측치 제외 된 데이터
```

데이터 시각화 하기

```python
plt.title('High')
plt.plot(result, 'r') # result 최고 기온 데이터를 빨간색 그래프로 그리기
plt.show
```

```python
# 그래프 크기 조절

plt.figure(figsize = (20, 2), dpi = 400)
plt.plot(result, 'r')
plt.show
```

```python
# 최고기온 값이 8월에 해당하는 값만 추출해서 그래프 시각화

```

```python
f = open('/content/seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

result = []

for row in data:
  if row[-1] != '':                     # 최고 기온 결측치 제거
    if row[0].split('-')[1] == '08':     # '-'를 기준으로 나누고 원하는 8월만 추출
      result.append(float(row[-1]))     # 실수형으로 저장

plt.figure(figsize = (20, 2), dpi = 200)
plt.title('08_High')
plt.plot(result, 'deeppink')
plt.show()
```

매년 생일 기준 최고 기온 그래프 그리기

```python
f = open('/content/seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

result = []

for row in data:
  if row[-1] != '':                                                       # 최고 기온 결측치 제거
    if row[0].split('-')[1] == '06' and row[0].split('-')[2] == '09':     # '-'를 기준으로 나누고 원하는 6월 9일 데이터 추출
      result.append(float(row[-1]))                                       # 실수형으로 저장

plt.figure(figsize = (20, 2), dpi = 200)
plt.title('08_High')
plt.plot(result, 'deeppink')
plt.show()
```

매년 생일 기준 최고 기온, 최저 기온 그래프 그리기

```python
f = open('/content/seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

high = []     # 최고 기온
low = []      # 최저 기온

for row in data:
  if row[-1] != '' and row[-2] != '':
    if row[0].split('-')[1] == '09' and row[0].split('-')[2] == '13':
      high.append(float(row[-1]))
    if row[0].split('-')[1] == '09' and row[0].split('-')[2] == '13':
      low.append(float(row[-2]))


plt.figure(figsize = (10, 2), dpi = 200)
plt.title('High & Low')
plt.plot(high, 'r', label = 'High')
plt.plot(low, 'b', label = 'Low')
plt.legend()
plt.show()
```

날짜 구간을 좁혀서 그래프 그리기

```python
f = open('/content/seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

high = []     # 최고 기온
low = []      # 최저 기온

for row in data:
  if row[-1] != '' and row[-2] != '':                                     # 결측치 제거
    if 1995 <= int(row[0].split('-')[0]):                                 # 1995년 부터
      if row[0].split('-')[1] == '09' and row[0].split('-')[2] == '13':   # 9월 13일
        high.append(float(row[-1]))
        low.append(float(row[-2]))

plt.figure(figsize = (5, 2), dpi = 150)
plt.title('내 생일의 최고, 최저 기온')
plt.plot(high, 'r', label = 'High')
plt.plot(low, 'b', label = 'Low')
plt.legend()
plt.show()
```

매년 생일 기준 최고 기온, 평균 기온, 최저 기온 그래프 그리기

```python
f = open('/content/seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

high = []     # 최고 기온
avg = []
low = []      # 최저 기온

for row in data:
  if row[-1] != '' and row[-2] != '' and row[-3] != '':                                     # 결측치 제거
    if 1995 <= int(row[0].split('-')[0]):                                 # 1995년 부터
      if row[0].split('-')[1] == '09' and row[0].split('-')[2] == '13':   # 9월 13일
        high.append(float(row[-1]))
        low.append(float(row[-2]))
        avg.append(float(row[-3]))

plt.figure(figsize = (5, 2), dpi = 150)
plt.title('내 생일의 최고, 최저 기온')
plt.plot(high, 'r', label = 'High')
plt.plot(avg, 'g', label = 'avg')
plt.plot(low, 'b', label = 'Low')
plt.legend()
plt.show()
```

히스토그램 hist() 함수 사용

```python
import matplotlib.pyplot as plt
import csv
import random
```

```python
plt.figure(dpi = 100) # 해상도
# 히스토그램 -> 도수 분포 -> 어떠한 자료에 속해 있는 값들을 수량으로 나타낸 것
plt.hist([1, 1, 2, 3, 4, 5, 6, 6, 7, 8, 10]) # hist(data(리스트)) 가로축은 계급(아이템), 세로축은 도수(횟수, 빈도수)
plt.show()
```

주사위 시뮬레이션 - 히스토그램 연습

```python
import random
```

```python
print(random.randint(1, 6)) # 1~6 사이의 값을 랜덤으로 출력
```

```python
# 주사위 5번 던져서 나온 데이터 출력
for i in range(5):
  print(random.randint(1, 6))
```

```python
# 주사위 5번 던져서 나온 데이터 최종 한 번에 출력
dice = []   # 주사위 리스트 변수 선언
for i in range(5):
  dice.append(random.randint(1, 6))

print(dice)
```

```python
# 주사위 5번 던져서 나온 데이터 값을 히스토그램으로 만들기
dice = []
for i in range(5):
  dice.append(random.randint(1, 6))

plt.figure(dpi = 150)
plt.title('Dice')
plt.hist(dice, bins=6)    # 주어진 bins값까지 구간을 생성해줌 ...안해주면 만약 6값이 안나왔을 때 5까지만 나옴
plt.show()
```

8월의 최고 기온 데이터 히스토그램으로 표현하기

```python
f = open('/content/seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

aug = []

for row in data:
  month = row[0].split('-')[1]  # 날짜 데이터에서 월 데이터 추출

  if row[-1] != '':
    if month == '09' and row[0].split('-')[2] == '13':
      aug.append(float(row[-1]))


plt.figure(dpi = 150)
plt.hist(aug, bins = 10, color = 'g')
plt.show()
```

1월과 8월의 데이터를 히스토그램으로 시각화 하기

```python
f = open('/content/seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

aug = []
april = []
jan = []

for row in data:
  month = row[0].split('-')[1]   # 날짜 데이터에서 월 데이터 추출

  if row[-1] != '':
    if month == '08':             # 8월
      aug.append(float(row[-1]))
    if month == '01':             # 1월
      jan.append(float(row[-1]))
    if month == '04':             # 4월
      april.append(float(row[-1]))


plt.figure(dpi = 150)
plt.hist(aug, bins = 300, color = 'r', label = 'Aug')
plt.hist(jan, bins = 300, color = 'b', label = 'Jan')
plt.hist(april, bins = 300, color = 'g', label = 'April')
plt.legend()
plt.show()
```

1월부터 12월까지 최고 기온 데이터 상자 그림으로 표현하기

```python
f = open('/content/seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

month = [[], [], [], [], [], [], [], [], [], [], [], []]

for row in data:
  if row[-1] != '':
    month[int(row[0].split('-')[1])-1].append(float(row[-1]))
    ## 월과 같은 번호의 인덱스에 월별 데이터 저장 (예: 1월 row[0].split('-')[1] -> month[0] row[0].split('-')[1]-1 , 2월 -> month[1])


plt.figure(dpi = 150)
plt.boxplot(month)
plt.show()
```

```python
f = open('/content/seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

day = []    # 일별 데이터를 저장할 리스트 day 변수 선언
for i in range(31):
  day.append([])    # day 리스트 내에 31개 리스트 생성 [[], [], [], ..., []]

for row in data:
  if row[-1] != '':
    if row[0].split('-')[1] == '08':
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))   # 8월 최고 기온 값 저장 day[0] -> 1일

plt.figure(figsize = (10, 2), dpi = 150)
plt.boxplot(day, showfliers  = False)    # 아웃라이어(이상치) 값 생략
plt.show()
```

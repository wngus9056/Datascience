```python
from collections import Counter    #카운터는 딕셔너리 스타일의 컬렉션에서 다른 컬렉션 안의 아이템 개수 셀 때 사용
phrase = 'a man a plan a canal panama'
cntr = Counter(phrase.split())
cntr.most_common(2)  # Default는 전부 숫자를 넣으면 그 수만큼 내림차순으로 출력
```

정규 표현식으로 패턴 매칭하기
re.split() 함수 단어 분석을 위한 기본적인 단어 토크나이저로 사용할 수 있음

```python
import re

re.split(r'\W', 'Hello, world!')   # r'\W' == 숫자, 알파벳 제외 한 개체 1개 기준으로 잘라라
```

```python
re.split(r'\W+', 'Hello, world!a!')  #인근에 위치한 비문자를 합친다.
```

```python
mo = re.match(r'\d+', '1234 12star with a number')
mo
```

```python
mo.group()
```

```python
print(re.match(r'\d+', 'star with a number'))       # 숫자가 없다.
print(re.match(r'\d+', 'star with123 a number'))    # 시작값부터 숫자가 없으면 못찾는다
```

```python
# search 문자열 일부분이 정규식과 매칭되는지 확인 (찾고자 하는 값이 앞부분에 없으면 이 함수 사용)
# search 는 해당 처음 값 하나만 출력
# findall 은 전부
# [a-z] 소문자들만 일치

print(re.search(r'[a-z]', '010010 Has at least one 010 letter 09301934').group())
print(re.search(r'[a-z]+', '010010 Has at least one 010 letter 09301934', re.I).group()) #re.I == 대소문자 구분안하고 가져와
```

```python
print(re.search(r'[A-Z]', '010010 Has at least one 010 letter 09301934').group())
print(re.search(r'[A-Z]+', '010010 has at least ONE 010 letter 09301934').group())
```

```python
# findall == 정규식에 부합하는 모든 문자열 찾기
print(re.findall(r'[a-z]+', '010010 Has at least one 010 letter 09301934'))
print(re.findall(r'[a-z]+', '010010 Has at least one 010 letter 09301934', re.I))
```

```python
print(re.findall(r'\d+', '010010 Has at least one 010 letter 09301934'))
print(re.findall(r'\D+', '010010 Has at least one 010 letter 09301934'))
```

```python
print(re.findall(r'^\d+', '010010 Has at least one 010 letter 09301934'))  # == not
```

```python
# sub : string으로 매칭되는 부분 문자열을 repl로 치환
re.sub(r'[a-z]+', '[...]', '010010 Has at least one 010 letter 09301934')
```

```python
re.sub(r'[a-z]+', '[...]', '010010 Has at least one 010 letter 09301934', flags = re.I)
```

```python
re.sub(r'[a-z]+', '[...]', '010010 Has at least one 010 letter 09301934',count=3, flags = re.I)
```

```python
# 010-abcd-0100  => abcd 값을 ****으로 치환
re.sub(r'[a-z]', '*', '010-abcd-0100')
```

```python
re.sub(r'[a-z]+', '****', '010-abcd-0100')
```

```python
# 숫자만 *로 치환
re.sub(r'\d', '*', '010-abcd-0100')
```

```python
import csv
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
```

```python
# 보람동 성별 데이터 수평 막대 그래프로 시각화 (항아리 그래프)
f = open('/content/gender.csv', 'r', encoding = 'cp949')
data = csv.reader(f)

male = []
female = []

for row in data:
  if '보람동' in row[0]:
    for i in row[3:104]:    # 남성 : 3 ~ 103번 데이터 슬라이싱 - 정수 데이터로 변환 - 리스트에 저장
      male.append(-int(i))
    for b in row[106:]:
      female.append(int(b))      # 여성 : 106 ~ 끝까지 데이터 슬라이싱 ''

plt.figure(dpi = 130)
plt.title('보람동 지역의 남녀 성별 인구 분포')
plt.barh(range(101), male, color = 'skyblue', label = '남성')
plt.barh(range(101), female, color = 'hotpink', label = '여성')
plt.legend()
plt.show
```

```python
f = open('/content/2012.csv', 'r', encoding = 'cp949')
data = csv.reader(f)

result2012 = []   # 순서를 위해 인덱스가 있는 리스트로 담는다.

for row in data:
  if '영등포동' in row[0]:
    for i in row[3:]:
      result2012.append(int(i.replace(',', '')))   # 2012년 보람동 데이터 정수로 변환 후 result2012 변수에 저장



f = open('/content/2021.csv', 'r', encoding = 'cp949')
data = csv.reader(f)

result2021 = []

for row in data:
  if '영등포동' in row[0]:
    for i in row[3:]:
      result2021.append(int(i.replace(',', '')))   # 2021년 보람동 데이터 정수로 변환 후 result2021 변수에 저장


plt.figure(dpi = 110)
plt.title('한솔동 연령별 인구')
plt.style.use('ggplot')
plt.plot(result2012, 'k', label = '2012')
plt.plot(result2021, 'r', label = '2021')
plt.legend()
plt.show()
```

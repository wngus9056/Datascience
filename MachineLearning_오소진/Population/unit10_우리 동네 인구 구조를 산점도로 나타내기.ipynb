```python
f = open('/content/gender.csv', 'r', encoding = 'cp949')
data = csv.reader(f)

m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
  if name in row[0]:
    for i in row[3:104]:    # 남성 : 3 ~ 103번 데이터 슬라이싱 - 정수 데이터로 변환 - 리스트에 저장
      m.append(int(i.replace(',', '')))
    for b in row[106:]:
      f.append(int(b.replace(',', '')))      # 여성 : 106 ~ 끝까지 데이터 슬라이싱 ''
    break   # 첫번 째 찾고 for문 중단

# 꺾은 선 그래프로 남여 인구 수 데이터 표현하기
plt.style.use('ggplot')
plt.figure(dpi = 110)
plt.plot(m, label = '남성')
plt.plot(f, label = '여성')
plt.legend()
plt.show()
```

연령별로 남성 인구수에서 여성 인구수를 뺀 차이의 막대그래프 그리기


```python
f = open('/content/gender.csv', 'r', encoding = 'cp949')
data = csv.reader(f)

result = [] # 남성 인구 - 여성 인구 값을 담을 리스트 변수 선언
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
  if name in row[0]:
    for i in range(3, 104):
      result.append(int(row[i].replace(',', '')) - int(row[i+103].replace(',', '')))
    break
      
# 꺾은 선 그래프로 남여 인구 수 데이터 표현하기
plt.style.use('ggplot')
plt.figure(dpi = 110)
plt.title(name+'지역 남여 인구 차이')
plt.bar(range(101), result)
plt.show()
```

점의 크기는 인구수의 제곱근 추가

```python
import math     # 제곱근을 사용하기 위해 임포트
f = open('/content/gender.csv', 'r', encoding = 'cp949')
data = csv.reader(f)

m = []
f = []
size = []   # 인구수 사이즈 리스트 변수 선언

name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
  if name in row[0]:
    for i in row[3:104]:    # 남성 : 3 ~ 103번 데이터 슬라이싱 - 정수 데이터로 변환 - 리스트에 저장
      m.append(int(i.replace(',', '')))
      size.append(math.sqrt(int(i.replace(',', ''))))   # 인구수 제곱근을 size
    for b in row[106:]:
      f.append(int(b.replace(',', '')))      # 여성 : 106 ~ 끝까지 데이터 슬라이싱 ''
    break   # 첫번 째 찾고 for문 중단

# 꺾은 선 그래프로 남여 인구 수 데이터 표현하기

plt.figure(dpi = 110)
plt.title('x = 남성  y = 여성')
plt.scatter(m,f,s = size, c = range(101), cmap = 'jet', alpha = 0.5)
plt.plot(range(max(m)), range(max(m)), 'g')   # 축은 x축을 기준으로
plt.colorbar()
plt.show()
```

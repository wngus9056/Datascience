```python
f = open('/content/age.csv', 'r', encoding = 'cp949')
data = csv.reader(f)

result = []   # 순서를 위해 인덱스가 있는 리스트로 담는다.

for row in data:
  if '보람동' in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',', '')))   # 인구수 데이터 정수로 변환 후 result 변수에 저장

print(result)
```

```python
plt.figure(dpi = 110)
plt.title('보람동 연령별 인구')
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

```python
f = open('/content/age.csv', 'r', encoding = 'cp949')
data = csv.reader(f)

result = []   # 순서를 위해 인덱스가 있는 리스트로 담는다.
result1 = []

for row in data:
  if '보람동' in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',', '')))   # 인구수 데이터 정수로 변환 후 result 변수에 저장
  if '아름동' in row[0]:
    for a in row[3:]:
      result1.append(int(a.replace(',', '')))

print(result)
print(result1)
```

```python
plt.figure(dpi = 110)
plt.title('세종시 연령별 인구')
plt.style.use('ggplot')
plt.plot(result, 'r', label = '보람동')
plt.plot(result1, 'b', label = '아름동')
plt.legend()
plt.show()
```
